## Helper Functions

Below are some helper functions to help get you started. They should look familiar from the lesson!

````python
import math

def grayscale(img):
    """Applies the Grayscale transform
    This will return an image with only one color channel
    but NOTE: to see the returned image as grayscale
    (assuming your grayscaled image is called 'gray')
    you should call plt.imshow(gray, cmap='gray')"""
    return cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)
    # Or use BGR2GRAY if you read an image with cv2.imread()
    # return cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    
def canny(img, low_threshold, high_threshold):
    """Applies the Canny transform"""
    return cv2.Canny(img, low_threshold, high_threshold)

def gaussian_blur(img, kernel_size):
    """Applies a Gaussian Noise kernel"""
    return cv2.GaussianBlur(img, (kernel_size, kernel_size), 0)

def region_of_interest(img, vertices):
    """
    Applies an image mask.
    
    Only keeps the region of the image defined by the polygon
    formed from `vertices`. The rest of the image is set to black.
    `vertices` should be a numpy array of integer points.
    """
    #defining a blank mask to start with
    mask = np.zeros_like(img)   
    
    #defining a 3 channel or 1 channel color to fill the mask with depending on the input image
    if len(img.shape) > 2:
        channel_count = img.shape[2]  # i.e. 3 or 4 depending on your image
        ignore_mask_color = (255,) * channel_count
    else:
        ignore_mask_color = 255
        
    #filling pixels inside the polygon defined by "vertices" with the fill color    
    cv2.fillPoly(mask, vertices, ignore_mask_color)
    
    #returning the image only where mask pixels are nonzero
    masked_image = cv2.bitwise_and(img, mask)
    return masked_image


def draw_lines(line_image, lines, color=[255, 0, 0], thickness=2):
    """
    The enhanced draw lines are written in the finding land lines pipeline directly.
    """

def hough_lines(img, rho, theta, threshold, min_line_len, max_line_gap):
    """
    `img` should be the output of a Canny transform.
        
    Returns an image with hough lines drawn.
    """
    lines = cv2.HoughLinesP(img, rho, theta, threshold, np.array([]), minLineLength=min_line_len, maxLineGap=max_line_gap)
    line_img = np.zeros((img.shape[0], img.shape[1], 3), dtype=np.uint8)
    draw_lines(line_img, lines)
    return line_img

# Python 3 has support for cool math symbols.

def weighted_img(img, initial_img, α=0.8, β=1., γ=0.):
    """
    `img` is the output of the hough_lines(), An image with lines drawn on it.
    Should be a blank image (all black) with lines drawn on it.
    
    `initial_img` should be the image before any processing.
    
    The result image is computed as follows:
    
    initial_img * α + img * β + γ
    NOTE: initial_img and img must be the same shape!
    """
    return cv2.addWeighted(initial_img, α, img, β, γ)
````

## Test Images

Build your pipeline to work on the images in the directory "test_images"  
**You should make sure your pipeline works well on these images before you try the videos.**

````python
import os
files = os.listdir("test_images/")

for file in files:
    if file[0:6] != "output":
        
        # Read in the image in the file.
        image = mpimg.imread("test_images/"+file)
        
        
        # STEP1. COLOR SELECTION
        
        # Grab the x and y size and make a copy of the image.
        ysize = image.shape[0]
        xsize = image.shape[1]
        
        image_select = np.copy(image)
        
        # Define the color selection criteria to cover yello line also.
        red_threshold = 150
        green_threshold = 100
        blue_threshold = 60
        rgb_threshold = [red_threshold, green_threshold, blue_threshold]
        
        # Identify pixels below the threshold
        thresholds = (image[:,:,0] < rgb_threshold[0]) \
                    | (image[:,:,1] < rgb_threshold[1]) \
                    | (image[:,:,2] < rgb_threshold[2])
        image_select[thresholds] = [0,0,0]
          
        
        # STEP2. REGION MASKING
        
        # Define a triangle region of interest
        left_bottom = [130, 539]
        right_bottom = [959, 539]
        apex = [475, 305]
        
        # Perform a linear fit (y=Ax+B) to each of the three sides of the triangle.
        # np.polyfit returns the coefficients [A, B] of the fit.
        # The arguments ((x1, x2), (y1, y2), 1). This means that A = (y2 - y1) / (x2 - x1) and B = y2 - A*x2.
        fit_left = np.polyfit((left_bottom[0], apex[0]), (left_bottom[1], apex[1]), 1)
        fit_right = np.polyfit((right_bottom[0], apex[0]), (right_bottom[1], apex[1]), 1)
        fit_bottom = np.polyfit((left_bottom[0], right_bottom[0]), (left_bottom[1], right_bottom[1]), 1)
        
        # Find the region inside the lines
        XX, YY = np.meshgrid(np.arange(0, xsize), np.arange(0, ysize))
        region_thresholds = (YY > (XX*fit_left[0] + fit_left[1])) & \
                            (YY > (XX*fit_right[0] + fit_right[1])) & \
                            (YY < (XX*fit_bottom[0] + fit_bottom[1]))
        
        # Region selection
        image_select[~region_thresholds] = [0,0,0]
        
        
        # STEP3. CANNY EDGE DETECTION
        
        # grayscale conversion
        gray = cv2.cvtColor(image_select, cv2.COLOR_RGB2GRAY)
        
        # Define a kernel size for Gaussian smoothing / blurring
        kernel_size = 3
        blur_gray = cv2.GaussianBlur(gray, (kernel_size, kernel_size), 0)
        
        # Canny function 
        # The algorithm will first detect strong edge (strong gradient) pixels above the high_threshold
        # and reject pixels below the low_threshold. The recommended a low to high ratio is 1:2 or 1:3.
        low_threshold = 50
        high_threshold = 150
        edges = cv2.Canny(blur_gray, low_threshold, high_threshold)  
        
        
        # STEP4. HOUGH TRANSFORM
        
        # Create a masked edges image using cv2.fillPoly()
        mask = np.zeros_like(edges)
        ignore_mask_color = 255
        
        # Define a four sided polygon to mask
        vertices = np.array([[(0,ysize),(450,300),(490,300),(xsize,ysize)]], dtype=np.int32)
        cv2.fillPoly(mask, vertices, ignore_mask_color)
        masked_edges = cv2.bitwise_and(edges, mask)
        
        # Define the Hough transform parameters
        rho = 1                        # distance resoultion in pixels of the Hough grid
        theta = np.pi/180              # angular resolution in radians of the Hough grid
        threshold = 30                # minimum number of votes (intersections in Hough grid cell)
        min_line_length = 2           # minimum number of pixels making up a line
        max_line_gap = 5              # maximum gap in pixels between connectable line segments
        line_image = np.copy(image)*0  # creating a blank to draw lines on
        
        # Run Hough on edge detected image
        # Output "lines" is an array containing endpoints of detected line segments
        lines = cv2.HoughLinesP(masked_edges, rho, theta, threshold, np.array([]), min_line_length, max_line_gap)

        # STEP5. IMPROVED DRAW LINES
        
        # Find the lines below an absolute value of a slope. 
        # This is slope threshold is determined by the triangular mask.
        # We have to choose the lower one between the left slope's absolute value and the right slope's absolute value.
        # abs(fit_left[0]) is the left slope's absolute value and abs(fix_rigth[0]) is the right slope's absolute value.
        slope_threshold = abs(fit_left[0]) if abs(fit_left[0]) < abs(fit_right[0]) else abs(fit_right[0])
        
        # Collect the lines which slope are above or equal to the slope threshold.
        # Filter which are left lines or right lines.
        left_lines = []
        right_lines = []
        
        for line in lines:
            x1, y1, x2, y2 = line[0]
            slope = (y2 - y1) / (x2 - x1)
            
            # Filter the slope
            if abs(slope) > slope_threshold:
                if slope < 0 :
                    left_lines.append(line)
                elif slope > 0 :
                    right_lines.append(line)
                    
        # Distribute x-axis and y-axis values in each left line or right line.
        left_x_vals = []
        left_y_vals = []
        right_x_vals = []
        right_y_vals = []

        for l_line in left_lines:
            x1, y1, x2, y2 = l_line[0]
            left_x_vals.extend([x1, x2])
            left_y_vals.extend([y1, y2])

        for r_line in right_lines:
            x1, y1, x2, y2 = r_line[0]
            right_x_vals.extend([x1, x2])
            right_y_vals.extend([y1, y2])
        
        # Obtain the linear polynominal slope and y-axis value.
        left_fit_line_param = np.polyfit(left_x_vals,left_y_vals,1)
        right_fit_line_param = np.polyfit(right_x_vals,right_y_vals,1)

        # Find the pairs two endpoints in each left side or right side.
        # I chooose the maximum value of y-axis and the correction value of apex[1] used in triangular masking 
        # as the two y-axis endpoints.
        y_endpoint_1 = int(ysize)
        y_endpoint_2 = int(apex[1] + 20)

        left_x_endpoint_1 = int((y_endpoint_1 - left_fit_line_param[1]) / left_fit_line_param[0])
        left_x_endpoint_2 = int((y_endpoint_2 - left_fit_line_param[1]) / left_fit_line_param[0])

        right_x_endpoint_1 = int((y_endpoint_1 - right_fit_line_param[1]) / right_fit_line_param[0])
        right_x_endpoint_2 = int((y_endpoint_2 - right_fit_line_param[1]) / right_fit_line_param[0])

        # Draw the left and right lines
        cv2.line(line_image,(left_x_endpoint_1,y_endpoint_1),(left_x_endpoint_2,y_endpoint_2),(255,0,0),10)
        cv2.line(line_image,(right_x_endpoint_1,y_endpoint_1),(right_x_endpoint_2,y_endpoint_2),(255,0,0),10)
        
        result = cv2.addWeighted(image, 0.8, line_image, 1, 0)
        plt.imshow(result)
        plt.show()
````

````python
def finding_lane_lines_pipeline(image):
    # STEP1. COLOR SELECTION
        
    # Grab the x and y size and make a copy of the image.
    ysize = image.shape[0]
    xsize = image.shape[1]
        
    image_select = np.copy(image)
        
    # Define the color selection criteria to cover yello line also.
    red_threshold = 150
    green_threshold = 100
    blue_threshold = 60
    rgb_threshold = [red_threshold, green_threshold, blue_threshold]
        
    # Identify pixels below the threshold
    thresholds = (image[:,:,0] < rgb_threshold[0]) \
                 | (image[:,:,1] < rgb_threshold[1]) \
                 | (image[:,:,2] < rgb_threshold[2])
    image_select[thresholds] = [0,0,0]
          
        
    # STEP2. REGION MASKING
        
    # Define a triangle region of interest
    left_bottom = [150, 539]
    right_bottom = [950, 539]
    apex = [475, 305]
        
    # Perform a linear fit (y=Ax+B) to each of the three sides of the triangle.
    # np.polyfit returns the coefficients [A, B] of the fit.
    # The arguments ((x1, x2), (y1, y2), 1). This means that A = (y2 - y1) / (x2 - x1) and B = y2 - A*x2.
    fit_left = np.polyfit((left_bottom[0], apex[0]), (left_bottom[1], apex[1]), 1)
    fit_right = np.polyfit((right_bottom[0], apex[0]), (right_bottom[1], apex[1]), 1)
    fit_bottom = np.polyfit((left_bottom[0], right_bottom[0]), (left_bottom[1], right_bottom[1]), 1)
        
    # Find the region inside the lines
    XX, YY = np.meshgrid(np.arange(0, xsize), np.arange(0, ysize))
    region_thresholds = (YY > (XX*fit_left[0] + fit_left[1])) & \
                        (YY > (XX*fit_right[0] + fit_right[1])) & \
                        (YY < (XX*fit_bottom[0] + fit_bottom[1]))
        
    # Region selection
    image_select[~region_thresholds] = [0,0,0]
        
        
    # STEP3. CANNY EDGE DETECTION
        
    # grayscale conversion
    gray = cv2.cvtColor(image_select, cv2.COLOR_RGB2GRAY)

    # Define a kernel size for Gaussian smoothing / blurring
    kernel_size = 3
    blur_gray = cv2.GaussianBlur(gray, (kernel_size, kernel_size), 0)
        
    # Canny function 
    # The algorithm will first detect strong edge (strong gradient) pixels above the high_threshold
    # and reject pixels below the low_threshold. The recommended a low to high ratio is 1:2 or 1:3.
    low_threshold = 50
    high_threshold = 150
    edges = cv2.Canny(blur_gray, low_threshold, high_threshold)  
        
        
    # STEP4. HOUGH TRANSFORM
        
    # Create a masked edges image using cv2.fillPoly()
    mask = np.zeros_like(edges)
    ignore_mask_color = 255
        
    # Define a four sided polygon to mask
    vertices = np.array([[(0,ysize),(450,300),(490,300),(xsize,ysize)]], dtype=np.int32)
    cv2.fillPoly(mask, vertices, ignore_mask_color)
    masked_edges = cv2.bitwise_and(edges, mask)
        
    # Define the Hough transform parameters
    rho = 1                        # distance resoultion in pixels of the Hough grid
    theta = np.pi/180              # angular resolution in radians of the Hough grid
    threshold = 30                 # minimum number of votes (intersections in Hough grid cell)
    min_line_length = 2           # minimum number of pixels making up a line
    max_line_gap = 5              # maximum gap in pixels between connectable line segments
    line_image = np.copy(image)*0  # creating a blank to draw lines on
        
    # Run Hough on edge detected image
    # Output "lines" is an array containing endpoints of detected line segments
    lines = cv2.HoughLinesP(masked_edges, rho, theta, threshold, np.array([]), min_line_length, max_line_gap)

    # STEP5. IMPROVED DRAW LINES
        
    # Find the lines below an absolute value of a slope. 
    # This is slope threshold is determined by the triangular mask.
    # We have to choose the lower one between the left slope's absolute value and the right slope's absolute value.
    # abs(fit_left[0]) is the left slope's absolute value and abs(fix_rigth[0]) is the right slope's absolute value.
    slope_threshold = abs(fit_left[0]) if abs(fit_left[0]) < abs(fit_right[0]) else abs(fit_right[0])
        
    for line in lines:
        x1, y1, x2, y2 = line[0]
        slope = (y2 - y1) / (x2 - x1)
            
        # Filter the slope
        if abs(slope) > slope_threshold:
            if slope < 0 :
                left_lines.append(line)
            elif slope > 0 :
                right_lines.append(line)
                    
    # Distribute x-axis and y-axis values in each left line or right line.
    left_x_vals = []
    left_y_vals = []
    right_x_vals = []
    right_y_vals = []

    for l_line in left_lines:
        x1, y1, x2, y2 = l_line[0]
        left_x_vals.extend([x1, x2])
        left_y_vals.extend([y1, y2])

    for r_line in right_lines:
        x1, y1, x2, y2 = r_line[0]
        right_x_vals.extend([x1, x2])
        right_y_vals.extend([y1, y2])
        
    # Obtain the linear polynominal slope and y-axis value.
    left_fit_line_param = np.polyfit(left_x_vals,left_y_vals,1)
    right_fit_line_param = np.polyfit(right_x_vals,right_y_vals,1)

    # Find the pairs two endpoints in each left side or right side.
    # I chooose the maximum value of y-axis and the correction value of apex[1] used in triangular masking 
    # as the two y-axis endpoints.
    y_endpoint_1 = int(ysize)
    y_endpoint_2 = int(apex[1] + 20)

    left_x_endpoint_1 = int((y_endpoint_1 - left_fit_line_param[1]) / left_fit_line_param[0])
    left_x_endpoint_2 = int((y_endpoint_2 - left_fit_line_param[1]) / left_fit_line_param[0])

    right_x_endpoint_1 = int((y_endpoint_1 - right_fit_line_param[1]) / right_fit_line_param[0])
    right_x_endpoint_2 = int((y_endpoint_2 - right_fit_line_param[1]) / right_fit_line_param[0])

    # Draw the left and right lines
    cv2.line(line_image,(left_x_endpoint_1,y_endpoint_1),(left_x_endpoint_2,y_endpoint_2),(255,0,0),10)
    cv2.line(line_image,(right_x_endpoint_1,y_endpoint_1),(right_x_endpoint_2,y_endpoint_2),(255,0,0),10)
    
    result = cv2.addWeighted(image, 0.8, line_image, 1, 0)
    
    return result
````
